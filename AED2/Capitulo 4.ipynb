A busca binária precisa de um vetor ordenado porque ela usa o valor do meio para decidir onde procurar. Se os dados não estiverem em ordem, ela pode procurar no lado errado e não encontrar o elemento, mesmo que ele exista.
# Exercício 3:

```python
def busca_binaria(vetor, alvo, esq, dir):
    if esq > dir:
        return -1

    meio = (esq + dir) // 2

    if vetor[meio] == alvo:
        return meio
    elif alvo < vetor[meio]:
        return busca_binaria(vetor, alvo, esq, meio - 1)
    else:
        return busca_binaria(vetor, alvo, meio + 1, dir)

#Exemplo de lista para a busca

lista = [10, 20, 30, 40, 50]

resultado = busca_binaria(lista, 30, 0, len(lista) - 1)
print(resultado)
```

```python
# Lista com 1000 números ordenados
lista = list(range(1, 1001))

# Pior caso (que não está na lista)
alvo = 2000


# Busca sequencial
def busca_sequencial(vetor, alvo):
    comparacoes = 0
    for i in range(len(vetor)):
        comparacoes += 1
        if vetor[i] == alvo:
            return comparacoes
    return comparacoes


# Busca binária
def busca_binaria(vetor, alvo):
    esq = 0
    dir = len(vetor) - 1
    comparacoes = 0

    while esq <= dir:
        comparacoes += 1
        meio = (esq + dir) // 2

        if vetor[meio] == alvo:
            return comparacoes
        elif alvo < vetor[meio]:
            dir = meio - 1
        else:
            esq = meio + 1

    return comparacoes


# Executando
comp_seq = busca_sequencial(lista, alvo)
comp_bin = busca_binaria(lista, alvo)

print("Busca sequencial:", comp_seq, "comparações")
print("Busca binária:", comp_bin, "comparações")
```

**Projeto 2: Busca Eficiente em um Índice de Livros**

**Contexto:** Você está trabalhando em um projeto para gerenciar uma biblioteca. As informações sobre os livros (título, autor, ano de publicação) estão armazenadas em uma lista, e essa lista é mantida ordenada alfabeticamente pelo título do livro.

**Problema:** Os usuários precisam encontrar rapidamente um livro específico sabendo seu título. Uma busca ineficiente pode dificultar a localização dos livros desejados. Indicação de como resolver: Desenvolva uma funcionalidade que permita aos usuários buscar um livro pelo título utilizando a busca binária. Ao receber o título do livro, sua função deve procurar no índice ordenado. Se o livro for encontrado, exiba suas informações (título, autor, ano). Caso contrário, informe que o livro não foi encontrado.

```python
#Realizando a busca binária
def busca_livros(livros, titulo):
    esq = 0
    di = len(livros) - 1
    while esq <= di:
        meio = (esq + di) // 2
        t_meio = livros[meio][0]
        if titulo == t_meio:
            return livros[meio]
        elif titulo > t_meio:
            esq = meio + 1
        else:
            di = meio - 1
    return None

#Lista dos livros que realizaremos a busca
livros=[
    ("A arte da guerra", "Sum Tzu", -500),
    ("Cem anos de solidão", "Gabriel Garcia Marquez", 1967),
    ("Dom casmurro", "Machado de Assis", 1899),
    ("O hobbit", "J R R Tolkien", 1937),
    ("Orgulho e preconceito", "Jane Austen", 1813)
]

print(busca_livros(livros, "O hobbit"))
```
