# Projeto 1: A Agenda de Contatos:
• Contexto: Você precisa criar um sistema simples para armazenar e gerenciar contatos (nome, telefone e e-mail). • Problema: Como você pode adicionar novos contatos, buscar um contato pelo nome e listar todos os contatos? • Indicação de como resolver: Pensar em uma estrutura que permita associar informações (telefone e e-mail) a um nome de forma eficiente. • Resolução do Projeto 1: Um dicionário em Python é uma excelente escolha para este projeto. A chave pode ser o nome do contato, e o valor pode ser outro dicionário contendo o telefone e o e-mail.

```python
agenda = {}

def adicionar_contato(nome, telefone, email):
    agenda[nome] = {"telefone": telefone, "email": email}
    print(f"Contato {nome} adicionado com sucesso!")

def buscar_contato(nome):
    if nome in agenda:
        info = agenda[nome]
        print(f"Nome: {nome} | Tel: {info['telefone']} | Email: {info['email']}")
    else:
        print("Contato não encontrado.")

def listar_contatos():
    print("\n--- Lista de Contatos ---")
    for nome, info in agenda.items():
        print(f"Nome: {nome} - Tel: {info['telefone']}")

adicionar_contato("vitoria", "3236-6897", "vitoria@email.com")
adicionar_contato("karina", "99845-8760", "karina@email.com")
buscar_contato("karina")
listar_contatos()
```

# Projeto 2: A Lista de Tarefas:
• Contexto: Você precisa criar uma lista de tarefas para a semana. • Problema: Como você pode adicionar novas tarefas, marcar tarefas como concluídas e visualizar as tarefas pendentes? • Indicação de como resolver: Pensar em uma forma de manter uma sequência de itens que podem ser adicionados e removidos. • Resolução do Projeto 2: Uma lista em Python é ideal para essa situação. Podemos adicionar novas tarefas ao final da lista, remover tarefas quando concluídas e simplesmente imprimir a lista para visualizar as tarefas pendentes.

```python
tarefas_pendentes = []

def adicionar_tarefa(tarefa):
    tarefas_pendentes.append(tarefa)
    print(f"Tarefa '{tarefa}' adicionada!")

def marcar_concluida(tarefa):
    if tarefa in tarefas_pendentes:
        tarefas_pendentes.remove(tarefa)
        print(f"Tarefa '{tarefa}' marcada como concluída e removida.")
    else:
        print("Tarefa não encontrada na lista.")

def visualizar_tarefas():
    print("\n--- Tarefas Pendentes ---")
    if not tarefas_pendentes:
        print("Nenhuma tarefa pendente.")
    for i, tarefa in enumerate(tarefas_pendentes, 1):
        print(f"{i}. {tarefa}")

adicionar_tarefa("Estudar Git e GitHub")
adicionar_tarefa("Revisar lógica proposicional")
visualizar_tarefas()
marcar_concluida("Estudar Git e GitHub")
visualizar_tarefas()
```
