```python
4. (PRÁTICA): Implemente uma função em Python chamada buscar_cliente(nome, lista_clientes) que
realiza busca sequencial em uma lista de nomes de clientes. Essa função deve retornar o índice onde o nome foi
encontrado ou -1 caso o nome não esteja na lista_clientes. Em seguida, escreva um pequeno código de
exemplo que utilize essa função para buscar por pelo menos um nome que exista e um que não exista na lista,
exibindo mensagens adequadas.
```

```python
def buscar_cliente(nome, lista_clientes):
    for i in range(len(lista_clientes)):
        if lista_clientes[i] == nome:
            return i
    return -1

clientes = ["Ana", "Bruno", "Caio"]
print(f"Busca 'Bruno': {buscar_cliente('Bruno', clientes)}")
print(f"Busca 'Zeca': {buscar_cliente('Zeca', clientes)}")
```

```python
def conta_ocorrencias(valor, lista):
    contador = 0
    for item in lista:
        if item == valor:
            contador += 1
    return contador

print(f"Ocorrências do 3: {conta_ocorrencias(3, [1, 3, 5, 3, 7])}")
```

```python
def remove_elemento(lista, valor):
    if not lista:
        return None
    
    for i in range(len(lista)):
        if lista[i] == valor:
            return lista.pop(i)
            
    return None

minha_lista = [10, 20, 30]
removido = remove_elemento(minha_lista, 20)
print(f"Removido: {removido}, Lista atual: {minha_lista}")
```

### Projeto 1: Encontrando seu Filme Favorito
Contexto: Imagine um sistema de recomendação de filmes simples onde os filmes são armazenados em uma lista não-ordenada.

Problema: Implementar uma função que permita ao usuário buscar um filme pelo título nessa lista.

Indicação de Solução: Crie uma lista de títulos de filmes e uma função que use a busca sequencial para encontrar um título inserido pelo usuário. Se o filme for encontrado, exiba uma mensagem informando o título e sua posição (índice) na lista. Se não for encontrado, exiba uma mensagem informando que o filme não está no catálogo.

```python
def buscar_filme(titulo_procurado, catalogo):
    for i in range(len(catalogo)):
        if catalogo[i].lower() == titulo_procurado.lower():
            return i
    return -1

meus_filmes = ["Matrix", "O Poderoso Chefão", "Interestelar", "Megamente"]

busca = input("Digite o título do filme que deseja buscar: ")

posicao = buscar_filme(busca, meus_filmes)

if posicao != -1:
    print(f"Sucesso! O filme '{meus_filmes[posicao]}' foi encontrado na posição {posicao}.")
else:
    print("O filme não está no catálogo.")
```

# Projeto 2: Verificando a Lista de Convidados
Contexto: Desenvolver um programa para gerenciar a lista de convidados de um evento.

Problema: Implementar uma funcionalidade que permita verificar se um nome específico está na lista de convidados.

Indicação de Solução: Crie uma lista de nomes de convidados. Implemente uma função que receba um nome como entrada e use a busca sequencial para verificar se esse nome existe na lista. Exiba uma mensagem indicando se o convidado está na lista ou não.

```python
def verificar_convidado(nome, lista_convidados):
    for convidado in lista_convidados:
        if convidado.lower() == nome.lower():
            return True
    return False

convidados = ["Sabrina", "Paty", "Karyna", "Mirella"]

nome_busca = input("Digite o nome do convidado: ")

if verificar_convidado(nome_busca, convidados):
    print("O convidado está na lista.")
else:
    print("O convidado não está na lista.")
```
