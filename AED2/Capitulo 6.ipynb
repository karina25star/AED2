```python
def decimal_para_hexa(n):
    if n == 0: return "0"
    hex_digits = "0123456789ABCDEF"
    stack = []
    while n > 0:
        stack.append(hex_digits[n % 16])
        n //= 16
    return "".join(reversed(stack))
```

```python
def balanceia_colchetes(expressao):
    stack = []
    pairs = {')': '(', '}': '{', ']': '['}
    for char in expressao:
        if char in "({[":
            stack.append(char)
        elif char in ")}]":
            if not stack or stack.pop() != pairs[char]:
                return False
    return not stack
```

## Projeto 1 – Histórico de Filtros em Dashboard de Dados

```python
class DashboardHistory:
    def __init__(self, estado_inicial):
        self.historico = []
        self.estado_atual = estado_inicial

    def aplicar_filtro(self, novo_filtro):
        self.historico.append(self.estado_atual)
        self.estado_atual = f"Dados filtrados por: {novo_filtro}"
        print(f"Filtro aplicado. Estado atual: {self.estado_atual}")

    def desfazer(self):
        if self.historico:
            self.estado_atual = self.historico.pop()
            print(f"Desfeito! Estado restaurado: {self.estado_atual}")
        else:
            print("Não há mais ações para desfazer.")

dash = DashboardHistory("Dados Brutos")
dash.aplicar_filtro("Vendas 2023")
dash.aplicar_filtro("Região Sul")
dash.desfazer()
```

# Projeto 2 – Calculadora de Expressões Aritméticas


```python
def calcular(expressao):
    def aplicar_op(valores, operadores):
        b = valores.pop()
        a = valores.pop()
        op = operadores.pop()
        if op == '+': valores.append(a + b)
        if op == '-': valores.append(a - b)
        if op == '*': valores.append(a * b)
        if op == '/': valores.append(a / b)

    precedencia = {'+': 1, '-': 1, '*': 2, '/': 2}
    valores = []
    operadores = []
    i = 0
    while i < len(expressao):
        if expressao[i] == ' ':
            i += 1; continue
        
        if expressao[i].isdigit():
            val = 0
            while i < len(expressao) and expressao[i].isdigit():
                val = (val * 10) + int(expressao[i])
                i += 1
            valores.append(val)
            continue
        
        if expressao[i] == '(':
            operadores.append('(')
        elif expressao[i] == ')':
            while operadores[-1] != '(':
                aplicar_op(valores, operadores)
            operadores.pop() # Remove '('
        else:
            while (operadores and operadores[-1] != '(' and 
                   precedencia[operadores[-1]] >= precedencia[expressao[i]]):
                aplicar_op(valores, operadores)
            operadores.append(expressao[i])
        i += 1

    while operadores:
        aplicar_op(valores, operadores)
    
    return valores[0]

resultado = calcular("6+9*(3-1)")
print(f"Resultado: {resultado}")
```
