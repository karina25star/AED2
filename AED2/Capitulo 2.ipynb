### Exercício 2:
Analise a complexidade da função encontrar_maximo(lista) que retorna o maior valor em uma lista: def encontrar_maximo(lista): nção de n?

```python
def encontrar_maximo(lista):
    valor_maximo = lista[0]
    for numero in lista:
        if numero > valor_maximo:
            valor_maximo = numero
            return valor_maximo
```


A lista de entrada tem n elementos. Qual a ordem de complexidade em função de n?

RESPOSTA: A complexidade é O(n), pois o algoritmo precisa percorrer todos os elementos da lista para encontrar o maior valor.

### Exercício 3:
Qual a complexidade da busca binária? Considere que temos uma lista ordenada de tamanho n e a função abaixo procura um elemento alvo nessa lista (retornando o índice ou -1 se não encontrar):

```python
def busca_binaria(lista, alvo):
    esquerda = 0
    direita = len(lista) - 1
    while esquerda <= direita:
        meio = (esquerda + direita) // 2
        if lista[meio] == alvo:
            return meio
        elif lista[meio] < alvo:
            esquerda = meio + 1
        else:
            direita = meio - 1
            return -1
```

Qual é a ordem de complexidade em função de n?

RESPOSTA: A complexidade é O(log n), pois a cada iteração o algoritmo reduz o espaço de busca pela metade.

### Exercício 4:
Considere o seguinte código que verifica se uma lista idades contém pelo menos duas ocorrências da menor idade presente na lista. Existem duas implementações diferentes para essa tarefa:

```python
def existe_2_menores_v1(idades):
    tamanho = len(idades)
    menor = 200 # supondo idades até 200
    # encontra a menor idade
    for i in range(tamanho):
        if idades[i] < menor:
            menor = idades[i]
    # conta ocorrências da menor idade
    cont = 0
    for i in range(tamanho):
        if idades[i] == menor:
            cont += 1
    return cont > 1


def existe_2_menores_v2(idades):
    idades.sort() # ordena a lista
    return idades[0] == idades[1]
```


Qual das duas funções tem menor complexidade de tempo? Justifique indicando as ordens de complexidade de cada uma.

RESPOSTA: A função existe_2_menores_v1 é mais eficiente, pois possui complexidade O(n), enquanto a função existe_2_menores_v2 possui complexidade O(n log n) devido à ordenação da lista.

### Exercício 5:
A função abaixo calcula o n-ésimo número de Fibonacci de forma recursiva (definindo fib(0)=0, fib(1)=1):

```python
def fib(n):
    if n < 2:
        return n
    else:
        return fib(n-1) + fib(n-2)
```

Qual é a complexidade de tempo dessa implementação em termos de n?

RESPOSTA: A complexidade é O(2ⁿ), pois a função realiza chamadas recursivas duplicadas, gerando um crescimento exponencial no número de operações.

### Projeto 1: Análise de Desempenho de Busca em Dados de Clientes
**Contexto:** Imagine que a Fatec Votorantim possui um sistema com informações sobre seus ex-alunos. Cada registro contém o nome, CPF e curso concluído. A coordenação precisa frequentemente buscar informações de um aluno específico utilizando o CPF como identificador único. O sistema atual realiza uma busca sequencial em uma lista de todos os ex-alunos.

**Problema:** Implemente duas funções em Python para buscar um ex-aluno pelo CPF em uma lista de dicionários (onde cada dicionário representa um ex-aluno). A primeira função (buscar_aluno_linear) deve usar uma busca linear, e a segunda função (buscar_aluno_eficiente) deve utilizar um dicionário para otimizar a busca, assumindo que os dados podem ser pré-processados.

**Indicação de como resolver:** Para buscar_aluno_linear, itere pela lista de ex-alunos e compare o CPF de cada um com o CPF buscado. Para buscar_aluno_eficiente, crie um dicionário onde a chave seja o CPF e o valor seja o dicionário do ex-aluno correspondente. Analise a complexidade de tempo de ambas as abordagens.

```python
# Lista de alunos
alunos = [
    {"nome": "Ana", "cpf": "111", "curso": "ADS"},
    {"nome": "Bruno", "cpf": "222", "curso": "Logística"},
    {"nome": "Carlos", "cpf": "333", "curso": "Gestão"}
]

# Busca linear
def buscar_aluno_linear(alunos, cpf):
    for aluno in alunos:
        if aluno["cpf"] == cpf:
            return aluno
    return None

# Criar índice (dicionário)
def criar_indice(alunos):
    indice = {}
    for aluno in alunos:
        indice[aluno["cpf"]] = aluno
    return indice

# Busca eficiente
def buscar_aluno_eficiente(indice, cpf):
    return indice.get(cpf, None)

# Testes
cpf_busca = "222"

print("Busca Linear:")
resultado = buscar_aluno_linear(alunos, cpf_busca)
if resultado:
    print(resultado)
else:
    print("Aluno não encontrado")

print("\nBusca Eficiente:")
indice = criar_indice(alunos)
resultado = buscar_aluno_eficiente(indice, cpf_busca)
if resultado:
    print(resultado)
else:
    print("Aluno não encontrado")
```

### Projeto 2: Recomendação de Disciplinas Baseada em Interesses
**Contexto:** A Fatec Votorantim oferece diversas disciplinas optativas. Para ajudar os alunos a escolherem, a coordenação quer criar um sistema simples de recomendação. Dado o histórico de disciplinas que um aluno já cursou, o sistema deve sugerir outras disciplinas que alunos com interesses semelhantes também cursaram.

**Problema:** Dada uma lista de históricos de disciplinas de alunos (onde cada histórico é uma lista de códigos de disciplinas) e o histórico de um aluno específico, implemente uma função em Python para encontrar disciplinas que outros alunos com pelo menos uma disciplina em comum cursaram, mas que o aluno original ainda não cursou.

**Indicação de como resolver:** Para cada outro aluno, compare seu histórico de disciplinas com o do aluno original. Se houver alguma disciplina em comum, adicione as disciplinas que o outro aluno cursou e que o aluno original não cursou a uma lista de recomendações. Analise a complexidade de tempo dessa abordagem.

```python
# Lista de históricos (cada lista representa um aluno)
historicos = [
    ["MAT1", "FIS1", "PROG1"],
    ["MAT1", "PROG1", "PROG2"],
    ["FIS1", "QUIM1"],
    ["MAT1", "FIS1", "QUIM1", "PROG2"]
]

# Histórico do aluno alvo
aluno = ["MAT1", "FIS1"]

def recomendar_disciplinas(historicos, aluno):
    recomendacoes = []

    for outro in historicos:
        # Verifica se tem alguma disciplina em comum
        comum = False
        for disc in outro:
            if disc in aluno:
                comum = True
                break

        # Se tiver em comum, adiciona novas disciplinas
        if comum:
            for disc in outro:
                if disc not in aluno and disc not in recomendacoes:
                    recomendacoes.append(disc)

    return recomendacoes


# Teste
resultado = recomendar_disciplinas(historicos, aluno)
print("Disciplinas recomendadas:", resultado)
```
